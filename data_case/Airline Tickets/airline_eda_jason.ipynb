# Import Dataset

## Convert Dataset to CSV

```python
import pandas as pd 
import os 

xlsx_file = 'airline_ticket_dataset.xlsx'
csv_file = 'airline_ticket_dataset.csv'

try: 
    df = pd.read_excel(xlsx_file, engine='openpyxl')
except ImportError: 
    print("Please install the openpyxl library: pip install openpyxl")

df.to_csv(csv_file, index=False)

data = pd.read_csv(csv_file)
data.describe()
```

## Display Dataset

Visualize image with histogram, and density plot. 

### Histogram

```python
import matplotlib.pyplot as plt

data.hist(figsize=(20, 10))
plt.tight_layout()
plt.show()
```

```python
data.plot.density()
plt.show()
```
